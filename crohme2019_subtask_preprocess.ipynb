```
STROKE_PAT= re.compile(r'<trace [^>]*>(.*?)</trace>',re.MULTILINE|re.DOTALL)
```

```
def pickup_traces(text):
  res = []
  pos = 0
  match = STROKE_PAT.search(text)
  while(match != None):
    res.append(match.group(1).strip("\n"))
    pos = match.end()
    match = STROKE_PAT.search(text, pos)
  return res
  
```

```
def file2dict(fpath, label_dict):
  with open(fpath, "r") as f:
    text = f.read()
  fkey = pickup_key(text)
  strokes = pickup_traces(text)
  return {"symbol": label_dict[fkey], "strokes":strokes, "path": fpath}
```

### Data cleansing

```
import re
```

```
EXCLAMATION_BEG_PAT=re.compile(r'^\\!(.*)')
EXCLAMATION_END_PAT=re.compile(r'(.*)\\!$')
SPACE_END_PAT=re.compile(r'(.*)\\ $')
```

```
def filter_trim_exclamation(target):
  tmp = re.sub(EXCLAMATION_BEG_PAT, r'\1',target)
  tmp = re.sub(EXCLAMATION_END_PAT, r'\1', tmp)
  tmp = re.sub(SPACE_END_PAT, r'\1', tmp)
  return tmp
  # return target.strip("\\!").rstrip("\\ ")
```

```
MATH_RM_PAT = re.compile(r'\\mathrm{([^}]*?)}')
```

```
def filter_mathrm(target):
  return re.sub(MATH_RM_PAT, r'\1', target)
```

```
MBOX_PAT = re.compile(r'\\mbox *{([^}]*?)}')
```

```
def filter_mbox(target):
  return re.sub(MBOX_PAT, r'\1', target)
```

```
LEFT_PAREN_PAT=re.compile(r'\\left *\(')
RIGHT_PAREN_PAT=re.compile(r'\\right *\)')
LEFT_SBRACKET_PAT=re.compile(r'\\left *\[')
RIGHT_SBRACKET_PAT=re.compile(r'\\right *\]')
```

```
def filter_left_right_paren(target):
  tmp = re.sub(LEFT_PAREN_PAT, '(', target)
  tmp =re.sub(RIGHT_PAREN_PAT, ')', tmp)
  tmp = re.sub(LEFT_SBRACKET_PAT, '[', tmp)
  tmp =re.sub(RIGHT_SBRACKET_PAT, ']', tmp)

  return tmp.replace("\\lbrack", "[").replace("\\rbrack", "]")
```

```
def filter_lg(target):
  return target.replace("&gt;", ">").replace("&lt;", "<").replace("\\gt ", "> ").replace("\\lt ", "< ")
```

```
filter_lg('\\lim_{x \\rightarrow -\\infty} p_2(x)&gt;0 \\!')
```

```
def filter_all(target):
  return filter_lg(filter_left_right_paren(filter_mbox(filter_mathrm(filter_trim_exclamation(target)))))
```

```
class Tokenizer:
  def __init__(self, one_tokens, multi_tokens):
    self.one_tokens = set(one_tokens)
    self.multi_tokens = multi_tokens
    self.target = ""
    self.pos = 0
    
  def set_target(self, target):
    self.pos=0
    self.target = target
  def skip_space(self):
    if self.is_end():
      return
    while (not self.is_end()) and self.target[self.pos] == ' ':
      self.pos += 1
  def go_next(self, token):
    self.pos += len(token)
    self.skip_space()
    return token
  def is_end(self):
    return len(self.target) == self.pos
  def tokenize_one(self):
    self.skip_space()
    multi = [tk for tk in self.multi_tokens if self.target.startswith(tk, self.pos)]
    if(len(multi) != 0):
      return self.go_next(multi[0])
    onech = self.target[self.pos]
    if(onech in self.one_tokens):
      return self.go_next(onech)
    self.unknown = onech
    raise Exception("Unknown token {}".format(onech))
  def tokenize_all_unsafe(self, target):
    self.set_target(target)
    self.skip_space()
    res = []
    while not self.is_end():
      res.append(self.tokenize_one())
    return True, res
  def tokenize_all(self, target):    
    try:
      return self.tokenize_all_unsafe(target)
    except:
      return False, None
      
```

### Setup stroke inupt

```
def strokearr2poslislis(strokearr):
  res = []
  
  for line in strokearr:
    res.append([(float(pair[0]),float(pair[1])) for pair in [pos.strip().split(" ") for pos in line.split(",")]])
  return res
```

```
def normalize_poslislis(poslislis):
    maxval = max(max(tup) for poslis in poslislis for tup in poslis)
    ratio = NORMALIZE_WIDTH/maxval
    return [(ratio*np.array(poslis)).astype(int) for poslis in poslislis]
    
```

```
def poslislis2vec(poslislis):
  """To list of 3D vector (x, y, TYPE).
  put (0, 0, INPUT_TYPE_SEP) between stroke.
  put (0, 0, INPUT_TYPE_END) at end.
  """
  res = None
  for one_stroke in poslislis:
    # onestoke (size, 2)
    widen = INPUT_TYPE_POINT*np.ones((one_stroke.shape[0], 3), dtype=int)
    widen[:, :-1] = one_stroke

    # Compare to None cause broadcast for np.array.
    if type(res) != np.ndarray:
      res = np.concatenate((widen, np.array([[0, 0, INPUT_TYPE_SEP]])))
    else:
      res = np.concatenate((res, widen, np.array([[0, 0, INPUT_TYPE_SEP]])))
  res[-1, 2] = INPUT_TYPE_END
  return res
```

```
def training_list_to_stroke_input_data(training_list):
  stroke_input_data = np.zeros(
    (len(training_list), STROKE_MAX_SEQ_LEN, INPUT_TYPE_DIM),
    dtype='int')
  for i, tup in enumerate(training_list):
    lislis = strokearr2poslislis(tup[3])
    lislis = normalize_poslislis(lislis)
    vec = poslislis2vec(lislis)
    stroke_input_data[i,0:vec.shape[0], : ] = vec
  return stroke_input_data
```
